```python
#Example of a gaussian distribution
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

# P,gaussian,2.798579071616348,1.0987784248566104e-07

mu = 2.798579071616348
variance = 1.0987784248566104e-07
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma),color='darkgreen')
plt.xlabel('Period [days]')
plt.ylabel('Frequency of samples')
plt.show()
```

```python
#Example of a uniform distribution
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

# rp,uniform,0.10522606200514388,0.12522606200514388,0

x=np.linspace(0.10522606200514388,0.12522606200514388,100)
y=np.ones(100)

plt.plot(x, y,color='darkgreen')
plt.xlabel('Radius')
plt.ylabel('Frequency of samples')
plt.show()
```
